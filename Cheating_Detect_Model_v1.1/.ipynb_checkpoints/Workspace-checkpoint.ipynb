```python
import librosa
import os
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
tf.compat.v1.set_random_seed(777)
```

```python
path_dir = "data"

X = []
Y = []
```

```python
sub_dirs = os.listdir(path_dir)

for sub_dir in sub_dirs:
    sub2_dirs = os.listdir(path_dir + '/' + sub_dir)
    for dir in sub2_dirs:
        
        print(dir)
        files = os.listdir(path_dir+ '/' + sub_dir + '/'+dir)
        for file_ in files:
            data = open(path_dir+ '/' + sub_dir + '/'+dir +'/' + file_, 'r')
            
            buffer = []
            line = data.readline()
            while line:
                mfcc = line.split()
                mfcc_f = list(map(float, mfcc))
                buffer.append(mfcc_f)
                line = data.readline()
            X.append(buffer)
            Y.append(dir)
```

```python
### X는 time step 10을 기준으로 짤라서 저장, 10보다 작은 데이터는 버리기 이때 Y 값도 같이 버려야 함 (Y 값과의 정확한 매핑이 중요) ###
timestep = 10

tempX = []
tempY = []

for i in range(len(X)):
    for j in range(0, len(X[i])-timestep , timestep):
        tempX.append( X[i][j:j+timestep] )
        tempY.append( Y[i] )
```

```python
tempX = np.array(tempX)
tempY = np.array(tempY)
```

```python
np.shape(tempX)
```

```python
print(tempY)
```

```python
label_y = []
for dir in tempY :
    if dir == 'Background' :
        label_y.append(0)
    elif dir == 'ballpen' :
        label_y.append(0)
    elif dir == 'breath' :
        label_y.append(0)
    elif dir == 'clonk' :
        label_y.append(0)
    elif dir == 'cough' :
        label_y.append(0)
    elif dir == 'eraser' :
        label_y.append(0)
    elif dir == 'snuffle' :
        label_y.append(0)
    elif dir == 'speech' :
        label_y.append(1)
    elif dir == 'whisper' :
        label_y.append(2)
    elif dir == 'ring' :
        label_y.append(3)
    elif dir == 'message' :
        label_y.append(4)
    elif dir == 'keyboard' :
        label_y.append(5)
    elif dir == 'mouse' :
        label_y.append(6)
```

```python
label_y = np.array(label_y)
```

```python
s = np.arange(tempX.shape[0])
np.random.shuffle(s)
```

```python
tempX = tempX[s]
label_y = label_y[s]
```

```python
print(np.shape(tempX))
print(np.shape(label_y))
```

```python
print(label_y)
```

```python
label_y = tf.keras.utils.to_categorical(label_y) # one-hot encoding
```

```python
print(label_y)
```

```python
print(np.shape(label_y))
print(np.shape(tempX))
```

```python
np.shape(tempX)
```

```python
print(tempX)
```

```python
x_train, x_test, y_train, y_test = train_test_split(tempX, label_y, test_size=0.3, random_state=777, stratify=label_y)
```

```python
x_train = np.array(x_train, dtype=np.float32)
x_test = np.array(x_test, dtype=np.float32)
```

```python
### Build model ###
input = tf.keras.layers.Input(shape=(10,12), dtype='float32')
net= tf.keras.layers.LSTM(units=32, activation='relu')(input) 
net = tf.keras.layers.Dense(units=32, activation='relu')(net)
net = tf.keras.layers.Dense(units=32, activation='relu')(net)
net = tf.keras.layers.Dense(units=32, activation='relu')(net)
net = tf.keras.layers.Dense(units=7, activation='softmax')(net)
model = tf.keras.models.Model(input, net)
```

```python
### Model Learning ###

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
#model.compile(loss=tf.keras.losses.CategoricalCrossentropy(), optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), metrics=[tf.keras.metrics.Accuracy()])

model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```
